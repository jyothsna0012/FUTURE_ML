```python
# ==========================================
# 1. Install Required Libraries
# ==========================================
!pip install pandas scikit-learn

# ==========================================
# 2. Import Libraries
# ==========================================
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# ==========================================
# 3. Create Dataset (Sample Data)
# ==========================================
data = {
    "ticket": [
        "Unable to login to my account",
        "Payment failed but money deducted",
        "App crashes when I open it",
        "Need refund for my order",
        "Password reset not working",
        "Website is very slow",
        "Charged twice for subscription",
        "Feature request for dark mode",
        "Account locked after multiple attempts",
        "Transaction failed but amount debited",
        "Error while uploading file",
        "Request to add new payment option"
    ],
    "category": [
        "Account", "Billing", "Technical", "Billing",
        "Account", "Technical", "Billing", "Feature",
        "Account", "Billing", "Technical", "Feature"
    ],
    "priority": [
        "High", "High", "High", "Medium",
        "Medium", "Low", "High", "Low",
        "High", "High", "Medium", "Low"
    ]
}

df = pd.DataFrame(data)

# ==========================================
# 4. Split Data
# ==========================================
X = df["ticket"]
y_category = df["category"]
y_priority = df["priority"]

X_train, X_test, y_cat_train, y_cat_test = train_test_split(
    X, y_category, test_size=0.2, random_state=42
)

_, _, y_pri_train, y_pri_test = train_test_split(
    X, y_priority, test_size=0.2, random_state=42
)

# ==========================================
# 5. Build Models (Pipelines)
# ==========================================
category_model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", RandomForestClassifier(n_estimators=100, random_state=42))
])

priority_model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", RandomForestClassifier(n_estimators=100, random_state=42))
])

# ==========================================
# 6. Train Models
# ==========================================
category_model.fit(X_train, y_cat_train)
priority_model.fit(X_train, y_pri_train)

# ==========================================
# 7. Evaluate Models
# ==========================================
cat_pred = category_model.predict(X_test)
pri_pred = priority_model.predict(X_test)

print("===== CATEGORY MODEL =====")
print("Accuracy:", accuracy_score(y_cat_test, cat_pred))
print(classification_report(y_cat_test, cat_pred))

print("\n===== PRIORITY MODEL =====")
print("Accuracy:", accuracy_score(y_pri_test, pri_pred))
print(classification_report(y_pri_test, pri_pred))

# ==========================================
# 8. Test with New Tickets
# ==========================================
new_tickets = [
    "I cannot access my account",
    "Refund my money immediately",
    "App not opening after update",
    "Please add dark theme",
    "Money deducted but order not placed"
]

cat_result = category_model.predict(new_tickets)
pri_result = priority_model.predict(new_tickets)

print("\n===== NEW PREDICTIONS =====")
for i in range(len(new_tickets)):
    print(f"\nTicket: {new_tickets[i]}")
    print(f"Predicted Category: {cat_result[i]}")
    print(f"Predicted Priority: {pri_result[i]}")

# ==========================================
# 9. Optional: Save Models
# ==========================================
import joblib

joblib.dump(category_model, "category_model.pkl")
joblib.dump(priority_model, "priority_model.pkl")

print("\nModels saved successfully!")
```

```python
# ==========================================
# 1. Install Libraries
# ==========================================
!pip install pandas scikit-learn joblib

# ==========================================
# 2. Import Libraries
# ==========================================
import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# ==========================================
# 3. Dataset (Expanded for Better Learning)
# ==========================================
data = {
    "ticket": [
        "Unable to login to my account",
        "Payment failed but money deducted",
        "App crashes when I open it",
        "Need refund for my order",
        "Password reset not working",
        "Website is very slow",
        "Charged twice for subscription",
        "Feature request for dark mode",
        "Account locked after multiple attempts",
        "Transaction failed but amount debited",
        "Error while uploading file",
        "Request to add new payment option",
        "Cannot update profile",
        "App freezing frequently",
        "Refund not processed yet",
        "Login OTP not received",
        "Payment page not loading",
        "Add multi-language support",
        "Server error 500 showing",
        "Unable to logout"
    ],
    "category": [
        "Account", "Billing", "Technical", "Billing",
        "Account", "Technical", "Billing", "Feature",
        "Account", "Billing", "Technical", "Feature",
        "Account", "Technical", "Billing", "Account",
        "Billing", "Feature", "Technical", "Account"
    ],
    "priority": [
        "High", "High", "High", "Medium",
        "Medium", "Low", "High", "Low",
        "High", "High", "Medium", "Low",
        "Medium", "High", "Medium", "High",
        "High", "Low", "High", "Medium"
    ]
}

df = pd.DataFrame(data)

# ==========================================
# 4. Train-Test Split
# ==========================================
X_train, X_test, y_cat_train, y_cat_test = train_test_split(
    df["ticket"], df["category"], test_size=0.2, random_state=42
)

_, _, y_pri_train, y_pri_test = train_test_split(
    df["ticket"], df["priority"], test_size=0.2, random_state=42
)

# ==========================================
# 5. Build Pipelines
# ==========================================
category_model = Pipeline([
    ("tfidf", TfidfVectorizer(stop_words='english')),
    ("clf", LogisticRegression(max_iter=200))
])

priority_model = Pipeline([
    ("tfidf", TfidfVectorizer(stop_words='english')),
    ("clf", LogisticRegression(max_iter=200))
])

# ==========================================
# 6. Train Models
# ==========================================
category_model.fit(X_train, y_cat_train)
priority_model.fit(X_train, y_pri_train)

# ==========================================
# 7. Evaluation
# ==========================================
print("=== CATEGORY MODEL ===")
print(classification_report(y_cat_test, category_model.predict(X_test)))

print("\n=== PRIORITY MODEL ===")
print(classification_report(y_pri_test, priority_model.predict(X_test)))

# ==========================================
# 8. Response Templates (Smart Replies)
# ==========================================
responses = {
    "Account": "We’re sorry you're facing account issues. Please try resetting your credentials or contact support.",
    "Billing": "We understand your billing concern. Our team is reviewing your transaction and will update you shortly.",
    "Technical": "Our technical team is working on resolving this issue. Please try again after some time.",
    "Feature": "Thank you for your suggestion! We’ll consider this feature in future updates."
}

# ==========================================
# 9. Prediction Function (Core System)
# ==========================================
def predict_ticket(ticket_text):
    category = category_model.predict([ticket_text])[0]
    priority = priority_model.predict([ticket_text])[0]

    # Confidence scores
    cat_conf = np.max(category_model.predict_proba([ticket_text]))
    pri_conf = np.max(priority_model.predict_proba([ticket_text]))

    response = responses.get(category, "Support will contact you soon.")

    return {
        "Ticket": ticket_text,
        "Category": category,
        "Priority": priority,
        "Category Confidence": round(cat_conf * 100, 2),
        "Priority Confidence": round(pri_conf * 100, 2),
        "Suggested Response": response
    }

# ==========================================
# 10. Test System
# ==========================================
test_tickets = [
    "I cannot login to my account",
    "Money deducted but payment failed",
    "App crashes after update",
    "Please add dark theme feature"
]

print("\n=== SAMPLE OUTPUT ===\n")

for t in test_tickets:
    result = predict_ticket(t)
    for k, v in result.items():
        print(f"{k}: {v}")
    print("-" * 50)

# ==========================================
# 11. Interactive Mode (User Input)
# ==========================================
while True:
    user_input = input("\nEnter a support ticket (or type 'exit'): ")
    if user_input.lower() == "exit":
        break

    result = predict_ticket(user_input)
    print("\n=== RESULT ===")
    for k, v in result.items():
        print(f"{k}: {v}")

# ==========================================
# 12. Save Models
# ==========================================
joblib.dump(category_model, "category_model.pkl")
joblib.dump(priority_model, "priority_model.pkl")

print("\nModels saved successfully!")
```
